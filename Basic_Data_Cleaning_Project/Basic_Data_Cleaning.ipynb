# Basic Data Exploration and Cleaning using Pandas

## Objective
Perform data exploration and cleaning on an Ecommerce Dataset using Pandas.

## Dataset Features
- Product Information
- Pricing Information
- Customer Ratings
- Seller Information
- Product Categories
- Product Reviews

```python
## Step 1: Import Required Libraries
import pandas as pd
import numpy as np

print("Libraries Imported Successfully")

# Step 2: Load the Dataset
df = pd.read_csv("archive/Combined_dataset.csv")

print("Dataset Loaded Successfully")

df[["initial_price", "final_price"]].head()

## Step 3: Display First Five Rows

df.head()

## Step 4: Display Last Five Rows

df.tail()

## Step 5: Dataset Shape

print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])

df.shape

## Step 6: Display Column Names

for column in df.columns:
    print(column)

## Step 7: Dataset Information

df.info()    

## Step 8: Statistical Summary

df.describe()

## Step 9: Statistical Summary Including Object Columns

df.describe(include="all")

## Step 10: Check Missing Values

missing_values = df.isnull().sum()

missing_values

## Step 11: Missing Value Percentage

missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage.sort_values(ascending=False)

## Step 12: Handle Missing Values

df_cleaned = df.copy()

object_columns = df_cleaned.select_dtypes(include="object").columns

df_cleaned[object_columns] = df_cleaned[object_columns].fillna("Unknown")

numeric_columns = df_cleaned.select_dtypes(include=np.number).columns

for col in numeric_columns:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

print("Missing Values Handled Successfully")

## Step 13: Verify Missing Values

df_cleaned.isnull().sum()

## Step 14: Check Duplicate Records

duplicate = df_cleaned.duplicated().sum()

print("Duplicate Rows:", duplicate)

## Step 15: Remove Duplicate Records

df_cleaned = df_cleaned.drop_duplicates()

print("Duplicates Removed Successfully")

## Step 16: Dataset Shape After Cleaning

df_cleaned.shape

## Step 17: Category Distribution

df_cleaned["category"].value_counts()

## Step 18: Top 10 Categories

df_cleaned["category"].value_counts().head(10)

## Step 19: Top Sellers
df_cleaned["seller_name"].value_counts().head(10)

## Step 20: Average Product Rating

df_cleaned["rating"].describe()

## Step 21: Top Rated Products

df_cleaned[
    ["title", "rating", "seller_name"]
].sort_values(
    by="rating",
    ascending=False
).head(10)

## Step 22: Price Analysis

df_cleaned[
    ["initial_price", "final_price"]
].head()

## Step 23: Cleaning Price Columns

df_cleaned["final_price"] = (
    df_cleaned["final_price"]
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)

## Step 24: Convert Price Column to Numeric

df_cleaned["final_price"] = (
    pd.to_numeric(
        df_cleaned["final_price"],
        errors="coerce"
    )
)

## Step 25: Verify Cleaned Price Columns

df_cleaned[
    [
        "initial_price",
        "final_price"
    ]
].head()

## Step 26: Create Derived Column (Discount Amount)

df_cleaned["discount_amount"] = (
    df_cleaned["initial_price"]
    -
    df_cleaned["final_price"]
)

## Step 27: Verify Discount Amount Column

df_cleaned[
    [
        "initial_price",
        "final_price",
        "discount_amount"
    ]
].head()

## Step 28: Highest Discount Products

df_cleaned[
    [
        "title",
        "discount_amount",
        "seller_name"
    ]
].sort_values(
    by="discount_amount",
    ascending=False
).head(10)

## Step 29: Filter Highly Rated Products

high_rating_products = (
    df_cleaned[
        df_cleaned["rating"] >= 4.5
    ]
)

high_rating_products.head()

## Step 30: Most Expensive Products

df_cleaned[
    [
        "title",
        "final_price"
    ]
].sort_values(
    by="final_price",
    ascending=False
).head(10)

## Step 31: Top Rated Products

df_cleaned[
    [
        "title",
        "rating"
    ]
].sort_values(
    by="rating",
    ascending=False
).head(10)

## Step 32: Save Cleaned Dataset

df_cleaned.to_csv(
    "cleaned_ecommerce_dataset.csv",
    index=False
)

print(
    "Cleaned Dataset Saved Successfully"
)
```
